#  Introduction

The healthcare dataset is a comprehensive collection of patient information,  facilitate research, analysis, providing valuable insights into various aspects of healthcare delivery. This dataset contains 15 essential variables that capture patient demographics, medical conditions, treatment details, and administrative information.This healthcare dataset serves as a comprehensive record of patient information for hospitals. It includes essential details for patient identification and treatment, allowing for better care management and streamlined operations. Each entry in the dataset consists of critical data points such as the patient's name, age, gender, blood type, medical condition, and insurance details. In addition to these basic attributes, the dataset tracks hospital-specific data like room number, admission dates, and billing amounts. Furthermore, it records vital medical information, including medications prescribed, test results, and treatments administered, along with key administrative details like doctor assignments, hospital information, and discharge dates.the dataset is designed to facilitate patient tracking, diagnosis, and treatment, ensuring efficient and personalized care delivery. this data can also be leveraged for hospital billing processes, insurance claims, and analytical studies aimed at improving healthcare services.

1. Patient Information: Name, Age, Gender, Blood Type
2. Medical Details: Medical Condition, Medication, Test Results
3. Hospitalization Data: Insurance, Billing Amount, Room Number, Admission Date, Discharge Date
4. Healthcare Provider Information: Doctor, Hospital

#### 1. which diseases have the highest number of patients admitted

#### 2. Which room numbers are most frequently occupied

#### 3. what is the average age of patients

```python
avg_age_male = df[df['Gender'] == 'Male']['Age'].mean()
avg_age_female = df[df['Gender'] == 'Female']['Age'].mean()
print("Average age of male patients:", avg_age_male)
print("Average age of female patients:", avg_age_female)
plt.figure(figsize=(8, 6))
plt.bar(['Male', 'Female'], [avg_age_male, avg_age_female], color=['#66b3ff', '#ffcc66'])
plt.title('Average Age of Male and Female Patients')
plt.xlabel('Gender')
plt.ylabel('Average Age')
plt.xticks(rotation=0)
plt.show()
```

The patient population is evenly split between genders, with each representing 50%.

#### 8. what is the averge billing amount for patients with insurance

```python
avg_billing_insured = df.loc[df['Insurance_provider'] == 'yes', 'Billing_amount'].mean()
avg_billing_by_provider = df.groupby('Insurance_provider')['Billing_amount'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Insurance_provider', y='Billing_amount', data=avg_billing_by_provider, palette='husl')
plt.title('Average Billing Amount by Insurance Provider')
plt.xlabel('Insurance Provider')
plt.ylabel('Average Billing Amount')
plt.xticks(rotation=90)
plt.show()
print("Average billing amount for patients with insurance:", avg_billing_insured)
```

#### 13. which month has the most patient admitted 

```python
df['Date_of_admission'] = pd.to_datetime(df['Date_of_admission'])
monthly_admissions = df['Date_of_admission'].dt.month.value_counts()
plt.figure(figsize=(6, 6))
plt.pie(monthly_admissions.values, labels=monthly_admissions.index, autopct='%1.1f%%',pctdistance=0.85, startangle=90)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Month with Most Patient Admissions')
plt.axis('equal')
plt.show()
print("Month with most patient admissions:\n", monthly_admissions)
```

The distribution of patient admissions across different months. It's evident that there's no significant variation in patient admissions throughout the year, with each month ranging between 7.7% and 8.7%. This suggests a consistent demand for healthcare services regardless of the time of year.

#### 16. how many patients have Hypertension & Asthma

#### 17. Which medical condition is reported mostly between age group of 30-60?

```python
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 60)]
grouped_df = filtered_df['Medical_condition'].value_counts()
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar')
plt.title('Medical Conditions Reported Between Age 30-60')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

The data reveals the various medical conditions among individuals aged 30-60,all affect a substantial number of people within this age group, with arthritis and diabetes leading the way. 

#### 18. what is the average age of cancer patients in young adults

```python
young_adults = df[(df['Age'] >= 15) & (df['Age'] <= 39) & (df['Medical_condition'] == 'Cancer')]
avg_age = young_adults['Age'].mean()
age_groups = young_adults['Age'].value_counts().sort_index()
plt.figure(figsize=(10, 8))
age_groups.plot(kind='pie', autopct='%1.1f%%')
plt.title('Age Distribution of Young Adult Cancer Patients')
plt.xlabel('')
plt.ylabel('')
plt.legend(title='Age', loc='upper right')
plt.show()
print("Average age of young adult cancer patients:", avg_age)
```

 The data shows that the most common age group for cancer diagnosis among young adults is 30 years old, representing 5.1% of the total cases.
